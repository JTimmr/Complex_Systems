### Testing the effect of $f$ on the model

```python
L = 50
remember_history = False
freeze_time_during_fire = True
instances = 1
Average_sparks = 500000

f = 19
timesteps = int(Average_sparks * f)

analysis_5 =  Analyse(L, f, freeze_time_during_fire, remember_history, timesteps, instances)
analysis_5.run_all()

analysis_5.find_best_fitting_distributions()
print(f'{analysis_5.best_fitting_distributions} distribution')


f = 76
timesteps = int(Average_sparks * f)

analysis_6 =  Analyse(L, f, freeze_time_during_fire, remember_history, timesteps, instances)
analysis_6.run_all()

analysis_6.find_best_fitting_distributions()
print(f'{analysis_6.best_fitting_distributions} distribution')

f = 305
timesteps = int(Average_sparks * f)

analysis_7 =  Analyse(L, f, freeze_time_during_fire, remember_history, timesteps, instances)
analysis_7.run_all()

analysis_7.find_best_fitting_distributions()
print(f'{analysis_7.best_fitting_distributions} distribution')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
r_squared_threshold = 0.95

label = f'f=19'
analysis_5.log_log_plot(ax, fig, color='black', label=label, scatter=False, r_squared_threshold=r_squared_threshold)

label = f'f=76'
analysis_6.log_log_plot(ax, fig, color='green',  label=label, scatter=False, r_squared_threshold=r_squared_threshold)

label = f'f=305'
title = 'Frequency fire sizes over all instances by fixing average number of sparks'
analysis_7.log_log_plot(ax, fig, color='orange',  title=title, label=label, scatter=False, r_squared_threshold=r_squared_threshold, show=True)
```
